```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
df=pd.read_csv('Business operation survey.csv')
df.head()
```

##  Viewing Data


```python
print(df.columns)
print()
print(df.sample(5))
print()
print(df.describe().sum())
print()
print(df.tail())
```

## Filtering the Survey Data

```python
print(df.query("level>1").tail())
print()
print(df[df["value"].between(200,500)])
print()
print()
```

## Survey Data Sorted by Level

```python
print(df.sort_values(by="level").head())
print()
print(df.sort_index().head())
```

##  Finding the Highest,lowest ,average and sum of survey values

```python
print("Sum of all values : ",df["value"].sum())
print("Minimum value : ",df["value"].min())
print("Maximum of Values : ",df["value"].max())
print("Average value : ",df["value"].mean())
```

# Visualization

## industry wise Business counts

```python
df["level"].value_counts().plot(kind="bar")
#df.head(5)
plt.title("Level of Survey")
plt.show()
```

# Size Class Distribution

```python
df['size'].value_counts().plot(kind='pie',autopct='%1.1f%%')
plt.ylabel('')
plt.show()
```

##  Levels of Survey

```python
df.groupby('level').size().plot(kind='line',marker='o',color='c')
plt.xticks(rotation=45)
plt.show()
```

## Industrial wise Value Distribution

```python
plt.hist(df['industry'],bins=10)
plt.title("Value Distribution")
plt.show
```

## Value vs Level

```python
plt.scatter(df['value'],df['level'])
plt.xlabel('value')
plt.ylabel('level')
plt.show()
```
